```python
import pandas as pd
import os
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Load the label CSV file
label_df = pd.read_csv('labels.csv')  # Update with the actual filename
len(label_df)
```

```python
label_df.columns = ['file','label']
label_df.head()
```

```python
folder_path = 'data'

# Initialize an empty list to store DataFrames
dfs = []

# Iterate over CSV files in the folder
for file_name in os.listdir('data'):
    #file_name = str(file_name_as_int).rjust(4,'0')
    # file_name is name of file without extension
    # Load CSV file
    
    df = pd.read_csv(folder_path + '\\' + file_name)

    file_name_as_int = int(file_name[:-4])
    # Extract the label from the label DataFrame based on the file name
    label = label_df.loc[label_df['file'] == file_name_as_int, 'label'].values[0]
    
    # Add a new column with the label
    df['label'] = [label]*len(df)

     # Add a new column with the file name
    file_name_wot_ext = str(file_name_as_int).rjust(4,'0')
    df['file'] = [file_name_wot_ext]*len(df)
    
    # Append the DataFrame to the list
    dfs.append(df)

# Merge all DataFrames into a single DataFrame
merged_df = pd.concat(dfs, ignore_index=True)

del merged_df['Unnamed: 0']

merged_df.head(5)
```

```python
# Convert time column to datetime format
merged_df['time'] = pd.to_datetime(merged_df['time'])

# Sort the DataFrame in ascending order of time
sorted_merged_df = merged_df.sort_values(by='time', ascending=True)

# Print the sorted DataFrame
sorted_merged_df.head(5)
```
